TRIANGULAR MEMBERSHIP

```python
import matplotlib.pyplot as plt

x=np.arange(0,10,0.1)

A=np.maximum(0,np.minimum(1,(x-2)/(6-2)))
B=np.maximum(0,np.minimum(1,(8-x)/(8-4)))
plt.xlabel("Universe of Discourse")
plt.ylabel("membership degree")
plt.title("triagular membership")
plt.plot(x,A,label='Set A')
plt.plot(x,B,label='Set B')
plt.legend()
plt.show()
```

```python
union=np.maximum(A,B)
plt.plot(x,A,label='Set A')
plt.plot(x,B,label='Set B')

plt.xlabel("Universe of Discourse")
plt.ylabel("membership degree")
plt.title("UNION")
plt.plot(x,union,label="UNION AUB",linestyle="dotted",color="red")
plt.legend()

plt.show()
```

```python
union=np.minimum(A,B)
plt.plot(x,A,label='Set A')
plt.plot(x,B,label='Set B')
plt.xlabel("Universe of Discourse")
plt.ylabel("membership degree")
plt.title("INTERSECTION")
plt.plot(x,union,label="Intersection of A and B",linestyle="dotted",color="red")
plt.legend()
plt.show()
```

```python
complement_A=1-A


plt.plot(x,A,label='Set A')

plt.xlabel("Universe of Discourse")
plt.ylabel("membership degree")
plt.title("complement of set A")
plt.plot(x,complement_A,label="complement of A",linestyle="dotted",color="green")
plt.legend()

plt.show()
```

```python
complement_B=1-B


plt.plot(x,B,label='Set B')

plt.xlabel("Universe of Discourse")
plt.ylabel("membership degree")
plt.title("complement of set A")
plt.plot(x,complement_B,label="complement of A",linestyle="dotted",color="green")
plt.legend()

plt.show()
```

```python
aggrement=(A+B)-(A*B)

plt.plot(x,A,label='Set A')
plt.plot(x,B,label='Set B')
plt.xlabel("Universe of Discourse")
plt.ylabel("membership degree")
plt.title("AGGREGATION")
plt.plot(x,aggrement,label="Aggregation AUB",linestyle="dotted",color="red")

plt.legend()

plt.show()
```

TRAPEZODIAL MEMBERSHIP

```python
x=np.arange(0,40,0.1)

cold=np.maximum(0,np.minimum(1,np.minimum((15-x)/(15-10),1)))
warm=np.maximum(0,np.minimum(1,np.minimum((x-10)/(20-10),(30-x)/(30-25))))
hot=np.maximum(0,np.minimum(1,np.maximum((x-25)/(30-25),0)))

plt.xlabel("Universe of Discourse")
plt.ylabel("membership degree")
plt.title("trapezodial membership")
plt.plot(x,cold,label='cold water')
plt.plot(x,warm,label='warm water')
plt.plot(x,hot,label='hot water')

plt.legend()
plt.show()
```
