```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 60, 0.1)                                    #speeds
Slow = fuzz.trapmf(x, [0, 10, 20, 25])
Medium = fuzz.trapmf(x, [20, 30, 40, 50])                    #ranges
Fast = fuzz.trapmf(x, [45, 50, 60, 70])

plt.figure(figsize=(10, 6))

plt.plot(x, Slow, 'b', linewidth=1.5, label='Slow ')
plt.plot(x, Medium, 'g', linewidth=1.5, label='Medium')
plt.plot(x, Fast, 'r', linewidth=1.5, label='Fast')
plt.ylabel('ranges')
plt.xlabel('speed')
plt.legend(loc='best')

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,40,0.1)
slow=np.maximum(0,np.minimum(1,np.minimum((15-x)/(15-10),1)))
medium=np.maximum(0,np.minimum(1,np.minimum((x-10)/(20-10),(30-x)/(30-25))))
fast=np.maximum(0,np.minimum(1,np.maximum((x-25)/(30-25),0)))
plt.figure(figsize=(12,6))
plt.plot(x,slow,label='Slow',linewidth=2)
plt.plot(x,medium,label='Medium',linewidth=2)
plt.plot(x,fast,label='Fast',linewidth=2)
plt.legend()
plt.grid(True)
plt.show()
```
